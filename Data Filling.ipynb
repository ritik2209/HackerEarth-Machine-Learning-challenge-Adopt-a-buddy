```python
def one_hot_encoding_top_x(df, variable, top_x_labels):
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode
    
    for label in top_x_labels:
        df[label] = np.where(df[variable]==label, 1, 0)
```

```python
def one_hot_encoding_topp_x(df2, variable, top_x_label):
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode
    
    for label in top_x_label:
        df2[label] = np.where(df2[variable]==label, 1, 0)
```

```python
one_hot_encoding_top_x(df, 'color_type', top_20_labels)
df.head()
```

```python
one_hot_encoding_topp_x(df2, 'color_type', top_20_label)
df.head()
```
